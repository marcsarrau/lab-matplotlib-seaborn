#### From the Fitbit data, we want to visually understand:

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fitdata = fitbit2.groupby(['Months_encoded', 'Work_or_Weekend'])['Steps'].median().reset_index()

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
x_fitdata = fitdata['Months_encoded']
y_fitdata = fitdata['Steps']
hue_fitdata = fitdata['Work_or_Weekend']
fig4, axes4 = plt.subplots(figsize=(12,4))
sns.barplot(x = x_fitdata, y = y_fitdata, hue = hue_fitdata)
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
active = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
x_col = 'Steps'
for i in active:
    fig5=plt.figure 
    ax5 = plt.gca()
    ax5.scatter(fitbit2[x_col], fitbit2[i])
    ax5.set_xlabel(x_col)
    ax5.set_ylabel(i)
    ax5.set_title("{} vs {}".format(x_col, i))
    plt.show()
    
```
